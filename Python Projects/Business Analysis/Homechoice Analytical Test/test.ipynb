# Read and Explore the Datasets

```python
import pandas as pd
```

```python
# Load dataset 1
FilePath = "Homechoice Analytical Test - 202401.xlsx"
loans_df = pd.read_excel(FilePath, sheet_name = "Dataset 1")
print("loans data shape:", loans_df.shape)
loans_df.head()
```

```python
# Load dataset 2
customers_df = pd.read_excel(FilePath, sheet_name = "Dataset 2")
print("Customer data shape:", customers_df.shape)
customers_df.head()
```

```python
# Merging the two datasets on ACCT_NO
merged_df = loans_df.merge(right=customers_df, how='inner', on="ACCT_NO")
print("Merged data shape:", merged_df.shape)
merged_df.head()
```

[QUESTION 1]

Create a report that shows the current outstanding balance & deposit account balance PER BRANCH
Which branch has the highest ratio of deposits compared to outstanding portfolio?

```python
branch_data = cleaned_df.groupby("BU_NO").agg(
    total_outstanding_balance=("CURRENT_BAL", "sum"),
    total_deposit_balance=("DEPOSIT_ACCOUNT", "sum")
).reset_index()

branch_data
```

```python
branch_data["deposit_to_outstanding_ratio"] = branch_data["total_deposit_balance"] / branch_data["total_outstanding_balance"]

highest_ratio_branch = branch_data.loc[branch_data["deposit_to_outstanding_ratio"].idxmax()]

print("\nBranch with the highest ratio of deposits to outstanding balance:")
highest_ratio_branch
```

[QUESTION 2]

Look at the disbursements for the month of May 2019 and give your thoughts on the products which made the biggest contribution to overall sales.

```python
may_2019_disbursements = cleaned_df[cleaned_df['DISBURSEMENT_DT'].dt.month == 5]
may_2019_disbursements = may_2019_disbursements[may_2019_disbursements['DISBURSEMENT_DT'].dt.year == 2019]

product_contributions = may_2019_disbursements.groupby('PROD_DESC')['DISBURSEMENT_AMT'].sum().reset_index()

total_disbursement = product_contributions['DISBURSEMENT_AMT'].sum()

product_contributions['contribution_percentage'] = (product_contributions['DISBURSEMENT_AMT'] / total_disbursement) * 100

product_contributions_sorted = product_contributions.sort_values(by='contribution_percentage', ascending=False)

print("Product contributions to disbursements for May 2019:")
product_contributions_sorted
```

#### [QUESTION 6]
	
Management need to have a summary of all products. You need to provide summary information that will allow them to answer the following questions:	
    Total Outstanding Balance per product?
    Frequency per product? 
    Summary information for Term, Disbursement Amt & Current Balance (For example: Average, Minimum, Maximum)


```python
cleaned_df['Term_Month'] = pd.to_numeric(cleaned_df['Term_Month'])
```

```python
product_summary = cleaned_df.groupby('PROD_DESC').agg(
    total_outstanding_balance=('CURRENT_BAL', 'sum'),
    frequency=('ACCT_NO', 'count'),
    avg_term_month=('Term_Month', 'mean'),
    min_term_month=('Term_Month', 'min'),
    max_term_month=('Term_Month', 'max'),
    avg_disbursement_amt=('DISBURSEMENT_AMT', 'mean'),
    min_disbursement_amt=('DISBURSEMENT_AMT', 'min'),
    max_disbursement_amt=('DISBURSEMENT_AMT', 'max'),
    avg_current_balance=('CURRENT_BAL', 'mean'),
    min_current_balance=('CURRENT_BAL', 'min'),
    max_current_balance=('CURRENT_BAL', 'max')
).reset_index()

# Print the summary report
print("Product Summary Information:")
product_summary
```
